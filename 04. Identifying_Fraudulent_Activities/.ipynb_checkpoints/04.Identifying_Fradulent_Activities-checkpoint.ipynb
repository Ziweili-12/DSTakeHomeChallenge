### Add country to the fraud table

```python
cou = []
for i in range(len(fraud)):
    ip_add = fraud.ip_address[i]
    data = ip[(ip['lower_bound_ip_address'] <= ip_add) &(ip['upper_bound_ip_address'] >= ip_add)]
    country = data.country.values
    
    if country.size >0:
        cou.append(country[0])
    else:
        cou.append("NA")
```

# Feature Engineering

```python
## check number of times that each devices may be used
dev_cnt = fraud.device_id.value_counts()
fraud['device_cnt'] = fraud.device_id.map(dev_cnt)
```

```python
## check number of times that each ip_address may be used
ip_cnt = fraud.ip_address.value_counts()
fraud['ip_cnt'] = fraud.ip_address.map(ip_cnt)
```

```python
## check each country's total users
coun_cnt = fraud.country.value_counts()
fraud['coun_cnt'] = fraud.country.map(coun_cnt)
```

```python
## check each source's total users
sour_cnt = fraud.source.value_counts()
fraud['sourc_cnt'] = fraud.source.map(sour_cnt)
```

```python
## check the seconds between singup and purchase time
fraud['time_interval'] = (fraud['purchase_time'] - fraud['signup_time']).dt.total_seconds()
```

```python
fraud['signup_year'] = fraud.signup_time.apply(lambda x: x.year)
fraud['signup_month'] = fraud.signup_time.apply(lambda x: x.month)
fraud['signup_day'] = fraud.signup_time.apply(lambda x: x.day)
```

```python
## because all the sign up happened in 2015
del fraud['signup_year']
```

```python
fraud['purchase_year'] = fraud.purchase_time.apply(lambda x: x.year)
fraud['purchase_month'] = fraud.purchase_time.apply(lambda x: x.month)
fraud['purchase_day'] = fraud.purchase_time.apply(lambda x: x.day)
```

```python
del fraud['purchase_year']
```

# Modeling
### Preparing data for modelling

```python
## Transform sex to numeric type
fraud['sex_10'] = fraud['sex'].apply(lambda x: 0 if x == 'F' else 1)
```

```python
data = pd.get_dummies(fraud, columns =['source','browser'], drop_first=True )
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit_transform(data['country'])
```

```python
data['country_code'] = le.fit_transform(data['country'])
```

```python
data.head(3)
```

```python
features = ['purchase_value', 'age','ip_address', 'device_cnt','ip_cnt','coun_cnt',
           'sourc_cnt', 'time_interval', 'signup_month', 'signup_day', 'purchase_month', 'purchase_day',
       'sex_10', 'source_Direct', 'source_SEO', 'browser_FireFox',
       'browser_IE', 'browser_Opera', 'browser_Safari','country_code']
```

```python
X = data[features]
y = data['class']
```

### Split data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
import xgboost as xgb
```

```python
train_matrix = xgb.DMatrix(X_train, y_train)
test_matrix = xgb.DMatrix(X_test)
```

```python
## Building parameters for cross-validation
params = {'objective':'binary:logistic', # logistic regression for binary classification,output probability
         'verbosity':1,
         'eta':0.2,
         'max_depth':6,
         'eval_metric':'auc',
          "early_stopping_rounds" : 30
         }
```

```python
xgb_result = xgb.cv(params, train_matrix,
                   metrics = params['eval_metric'],
                   verbose_eval = True,
                   early_stopping_rounds = params['early_stopping_rounds'],
                   nfold = params.get('nfold',3),
                   )
```

```python
xgb_result
```

```python
best_tree = xgb_result.shape[0]
best_tree
```

# Model interpretation
## ROC_Curve

```python
from sklearn.metrics import roc_curve
```

```python
## Using test set to plot the roc curve
X_train2, X_valid, y_train2, y_valid = train_test_split(X_train, y_train, 
                                                            test_size=0.3, random_state=42)
train_matrix2 = xgb.DMatrix(X_train2, y_train2)
valid_matrix2 = xgb.DMatrix(X_valid, y_valid)
    
xgb2 = xgb.train(params, train_matrix2, best_tree,[(train_matrix2,'train'), (valid_matrix2,'valid')])
    
y_pred = xgb2.predict(valid_matrix2, ntree_limit = best_tree)
y_pred2 = xgb2.predict(train_matrix2, ntree_limit= best_tree)   

fpr, tpr, threshold = roc_curve(y_valid, y_pred)
fpr2, tpr2, threshold2 = roc_curve(y_train2, y_pred2)
    
df = pd.DataFrame({'FPR':fpr, 'TPR':tpr, "Threshold":threshold})    
df2 = pd.DataFrame({'FPR':fpr2, 'TPR':tpr2, "Threshold":threshold2})
```

```python
# test
df.head(3)
```

```python
#train
df2.head(3)
```

```python
plt.figure(dpi=(100))
plt.plot(df.FPR,df.TPR, color='blue',label = 'Test AUC')
plt.plot(df2.FPR, df2.TPR, color='red',label = 'Train AUC')
plt.plot(df2.FPR, df2.FPR,'k--')
plt.xlabel("False Positive")
plt.ylabel("True Positive")
plt.legend()
```

From the plot, we can see that if we want to increase the True Positive, then we need to increase the threshold. However, increasing threshold means we will label more emails as spam, some of them will be True Spam, while others will be false spam. 

If we want to minimize the False Positive, then we need to decrease the threshold, like 0-0.05.  In this way, the false positive will be close to zero, and true positive will be around 0.05.

Here, since this step is just the first step to decide whether the email is a smap or not, we need experts to make the final call.But we might meet some cases:
1. if the model label this email is spam but it is not, some people will hand this manually.
2. but if the model label this email is not spam but it is, then the company will directly lose money.

In order to fix this problem, we need to label emails as spam as much as possible, so the 'false posiive'  threshold need to be smaller.

# How to use this model?

From the product prospective, since we are using this model to predict the probability, so we can set two level as normal and supicious. For every purchase, there will be a probability burying in the user account and to the office.
For example:
1. If the probability is below the normal level, then we believe is this not a spam, everything will run normally, and no notifications will show up.
2. If the probability is between normal and supicious level, then customer will need additional steps to confirm their identity, like from messages or voice-call, or email. Aftering confirming, they can continue the following steps.
3. If the probability is equal or bigger than supicious level, then the order will not be made successfully if they cannot confirm their identity, and the technical person will monitor this activity to do more investigation to make sure if this is a spam or not.
