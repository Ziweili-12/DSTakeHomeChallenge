# Analysis

```python
# define a function to calculate the conversion rate
def conversion_rate(df):
    
    n_length = len(df)
    
    home = df.page_home.count()
    search = df.page_search.count()
    payment = df.page_payment.count()
    confi = df.page_confirmation.count()
    
    n_convert = [home, search, payment, confi]
    n_left = [home - home, home - search, search - payment, payment - confi]
    
    conv_rate = pd.array([home/home, search/home, payment/search, confi/payment])
    leave_rate = 1-conv_rate
    
    df2 = pd.DataFrame({'stage':['home','search','page','confirm'],
                        'n_convert':n_convert,
                        'n_left': n_left,
                        'conversion_rate':conv_rate,
                        'leave_rate': leave_rate})
    return df2
```

### Overall

```python
## What is the overall conversion rate
conversion_rate(data)
```

### Device

```python
## What is each device's conversion rate in each stage
rate = data.groupby(['device']).apply(conversion_rate)
```

```python
rate
```

We can see that overall conversion rate is 0.074959. But conversion rate of mobile is 0.1, which is higher than Desktop(0.049834).

```python
rate2 = rate.reset_index()
```

```python
rate2['level_1'] = rate2['level_1'].astype('str')
```

```python
rate2
```

```python
plt.figure(dpi = (100))
sns.lineplot(x = 'level_1', y = 'conversion_rate', data=rate2, hue='device')
```

As we can see from each stage's conversion rate plot, from home-page to search page, the rate drops significantly. And the line of mobile is always above desktop's, meaning the conversion rate of mobile is always bigger than mobile's.
